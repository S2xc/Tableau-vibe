**Что такое Calculated Fields и зачем они нужны**

В Tableau есть много способов работать с данными — фильтры, сортировка, группировки и т.д. Но чтобы создавать по-настоящему гибкие и умные визуализации, этого недостаточно.
Именно для этого и существуют **вычисляемые поля (Calculated Fields)**.

Проще говоря, **Calculated Field** — это формула, которую ты сам придумываешь, чтобы получить нужный результат из уже существующих данных.

**Зачем они нужны?**
Вот несколько реальных примеров:
1) Хочешь показать **скидку в процентах**, а в данных есть только цена и цена со скидкой — пишешь формулу.

У пользователя ~Ondrej Mészáros~ возникла проблема:
Он хочет посчитать **итоговую скидку на уровне категорий товаров**, но его формула считает скидку **для каждой строки**, а затем **просто суммирует** — что даёт **неправильный итог** при агрегации. Ему нужно пересчитать **скидку от общей суммы**, а не сумму индивидуальных скидок.

(Ссылка на вопрос: https://community.tableau.com/s/question/0D58b00009Wp49bCAB/how-to-calculate-total-discount-on-aggregate-categories)

2) Нужно выделить **пользователей, у которых больше 5 заказов** — пишешь условие.
3) Хочешь сравнить продажи **в этом месяце и в прошлом** — снова нужна формула.
4) Не хватает группировки по логике типа “если A, то X, иначе Y” — это тоже легко сделать с помощью Calculated Field.

**Почему это важно?**
Формулы дают тебе **контроль**.
Ты сам решаешь, какие данные показывать, как их считать и как отображать.
Без них ты ограничен только тем, что уже есть в таблице. С ними — ты сам создаёшь нужные тебе поля, условия и логику.

### Синтаксис формул: поля данных, операторы (+, -, *, /), функции
Цель**: Научиться понимать и писать простые формулы в Tableau, используя поля данных, операторы и функции.

## Теория
**Что такое синтаксис формул в Tableau?**
Формула в Tableau — это инструкция, которая говорит программе, как выполнять вычисления с данными. Формулы создаются в **Calculated Fields** (вычисляемых полях) и состоят из трех основных элементов:  
* **Поля данных**: Это столбцы из вашего датасета, например, [Sales] (продажи) или [Order Date] (дата заказа).  
* **Операторы**: Математические символы, такие как + (сложение), - (вычитание), * (умножение), / (деление).  
* **Функции**: Готовые команды Tableau, которые выполняют определенные действия, например, SUM (сумма) или ROUND (округление).

⠀**Зачем это нужно?**
Формулы позволяют создавать новые показатели, которые не существуют в исходных данных. Например, вы можете рассчитать прибыль ([Sales] - [Cost]) или процент скидки ([Discount] / [Sales] * 100).  

⠀**Правила написания формул**:  
* Поля данных пишутся в квадратных скобках, например, [Sales].  
* Операторы используются как в математике: 2 + 2, [Sales] * [Quantity].  
* Функции имеют формат: НАЗВАНИЕ(аргумент), например, SUM([Sales]).  
* Используйте автодополнение в редакторе, чтобы избежать ошибок.

⠀
## Примеры формул
* **Пример 1: Расчет прибыли**  
  * Формула: [Sales] - [Cost] // Рассчитывает прибыль  
  * Что делает: Вычитает затраты ([Cost]) из продаж ([Sales]).  
  * Результат: Если [Sales] = 1000 и [Cost] = 600, то прибыль = 400.
* **Пример 2: Процент скидки**  
  * Формула: ([Discount] / [Sales]) * 100 // Процент скидки  
  * Что делает: Делит сумму скидки на продажи и умножает на 100, чтобы получить процент.  
  * Результат: Если [Discount] = 50 и [Sales] = 1000, то процент скидки = 5%.
* **Пример 3: Сумма продаж с функцией**  
  * Формула: SUM([Sales]) // Суммирует продажи  
  * Что делает: Суммирует все значения в поле [Sales] для выбранного уровня агрегации (например, по категории).  
  * Результат: Если в категории три заказа с продажами 200, 300 и 500, то SUM([Sales]) = 1000.

*Цель: Понять, какие типы данных используются в Tableau, как они влияют на формулы и как их преобразовывать для вычислений.*

## Теория
**Что такое типы данных в Tableau?**
Тип данных определяет, какого рода информация содержится в поле вашего датасета, и как Tableau будет с ней работать. Когда вы пишете формулы, важно знать тип данных, чтобы использовать правильные функции и операторы. В Tableau есть четыре основных типа данных, которые часто используются в вычислениях:  
* **Числовые**: Числа, с которыми можно выполнять математические операции (например, продажи, количество).  
* **Строковые**: Текстовые данные, такие как имена, категории или описания.  
* **Даты**: Значения, связанные с датами и временем (например, дата заказа).  
* **Логические**: Значения "истина" (TRUE) или "ложь" (FALSE), используемые для условий.

⠀**Зачем это нужно?**
Тип данных определяет, какие операции можно выполнять. Например, числа можно складывать, строки — объединять, а даты — сравнивать. Если тип данных неверный, формула может дать ошибку. Зная типы данных, вы сможете преобразовывать их (например, число в строку) и создавать точные вычисления.
**Как проверить тип данных?**
В панели данных Tableau каждое поле имеет иконку:  
* **Числа**: Значок # (например, [Sales], [Quantity]).  
* **Строки**: Значок Abc (например, [Customer Name], [Category]).  
* **Даты**: Значок календаря (например, [Order Date]).  
* **Логические**: Значок T/F (например, результат условия).

## Примеры формул
* **Числовые данные**  
  * **Пример**: [Sales] * [Quantity] // Рассчитывает общую стоимость.  
  * **Что делает**: Умножает числовое поле [Sales] (продажи) на [Quantity] (количество).  
  * **Результат**: Если [Sales] = 100 и [Quantity] = 5, то результат = 500.  
  * **Особенность**: Числовые данные подходят для операций +, -, *, / и функций, таких как SUM, AVG, ROUND.
* **Строковые данные**  
  * **Пример**: CONCAT([First Name], " ", [Last Name]) // Создает полное имя.  
  * **Что делает**: Объединяет строковые поля [First Name] и [Last Name] с пробелом.  
  * **Результат**: Если [First Name] = "Анна" и [Last Name] = "Иванова", то результат = "Анна Иванова".  
  * **Особенность**: Строки используются с функциями CONCAT, LEFT, RIGHT, CONTAINS.
* **Данные типа "Дата"**  
  * **Пример**: DATENAME('month', [Order Date]) // Извлекает название месяца.  
  * **Что делает**: Возвращает название месяца из поля [Order Date].  
  * **Результат**: Если [Order Date] = 2025-06-21, то результат = "June".  
  * **Особенность**: Даты работают с функциями DATEPART, DATEDIFF, DATEADD.
* **Логические данные**  
  * **Пример**: [Sales] > 1000 // Проверяет, превышают ли продажи 1000.  
  * **Что делает**: Возвращает TRUE, если [Sales] > 1000, или FALSE, если нет.  
  * **Результат**: Если [Sales] = 1200, то результат = TRUE.  
  * **Особенность**: Логические значения используются в условиях (IF, IIF) и фильтрах.
* **Преобразование типов данных**  
  * **Пример**: STR([Sales]) // Преобразует число в строку.  
  * **Что делает**: Преобразует числовое поле [Sales] в строку для текстовых операций.  
  * **Результат**: Если [Sales] = 1000, то результат = "1000".  
  * **Пример преобразования**: INT([String Number]) // Преобразует строку "123" в число 123.

## Рекомендации для новичков
* **Проверяйте тип данных перед созданием формулы**: В панели данных посмотрите на иконку поля ([Sales] — #, [Category] — Abc). Если тип неверный, формула может не сработать.  
* **Используйте правильные функции**: Например, SUM не работает со строками, а CONCAT — с числами без преобразования.  
* **Преобразуйте типы, если нужно**: Используйте STR, INT, DATEPARSE для изменения типов данных.  
* **Добавляйте комментарии**: Например, // Преобразует число в строку для ясности формулы.  

### Форматирование формул: использование комментариев (// и /* */) и отступов для читаемости
**Теория
Что такое форматирование формул?**
Форматирование формул — это способ сделать их понятными и удобными для чтения. В Tableau вы можете использовать два типа комментариев — однострочные (//) и многострочные (/* */), а также отступы, чтобы структурировать формулы. Это помогает вам и вашей команде быстро понять логику вычислений, особенно если формулы сложные.
**Зачем это нужно?**  
* Читаемые формулы легче проверять и исправлять.  
* Комментарии поясняют назначение формулы, что полезно при возвращении к ней позже.  
* Хорошо оформленные формулы упрощают совместную работу с коллегами.

⠀**Как форматировать формулы?**  
* **Однострочные комментарии (//)**: Используйте // для коротких пояснений в одной строке. Всё, что написано после //, игнорируется Tableau.  
* **Многострочные комментарии (/* */)**: Используйте /* для начала комментария и */ для его окончания. Это удобно для длинных пояснений или временного отключения части формулы.  
* **Отступы**: Добавляйте пробелы или переносы строк, чтобы разделить части сложной формулы и сделать её более читаемой.  
* **Логичное именование**: Называйте вычисляемые поля понятно, например, "Profit_Calculation" вместо "Calculation1".

⠀**Где это делается?**
В редакторе вычислений:  
* В панели данных щелкните правой кнопкой мыши → **Create Calculated Field**.  
* Пишите формулу, добавляя комментарии (// или /* */) и отступы для ясности.

⠀**Примеры**
* **Простая формула с однострочным комментарием**  
  * **Формула**:   [Sales] - [Cost] // Рассчитывает прибыль
  * **Что делает**: Вычитает затраты из продаж и добавляет комментарий для ясности.  
  * **Результат**: Если [Sales] = 1000 и [Cost] = 600, то результат = 400.
* **Сложная формула с многострочными комментариями и отступами**  
  * **Формула**:  
  
  /* 
  Рассчитывает процент скидки, 
  возвращает 0, если продаж нет, 
  чтобы избежать деления на ноль 
  */
  IF [Sales] > 0 THEN
  ([Discount] / [Sales]) * 100
  ELSE
  0
  END
  
  * **Что делает**: Вычисляет процент скидки, возвращает 0 при нулевых продажах. Многострочный комментарий поясняет логику, а отступы делают формулу читаемой.  
  * **Результат**: Если [Discount] = 50 и [Sales] = 1000, то результат = 5%.
* **Логичное имя поля с комментарием**  
  * **Название поля**: "Discount_Percentage"  
  * **Формула**:   [Discount] / [Sales] * 100 /* Вычисляет процент скидки */
  * **Что делает**: Название поля сразу говорит о его назначении, а комментарий поясняет формулу.

⠀**Рекомендации для новичков**
* **Используйте // для коротких пояснений**: Например, // Сумма продаж.  
* **Используйте /* */ для длинных пояснений**: Подходит для описания сложной логики или временного отключения части формулы.  
* **Добавляйте отступы в сложных формулах**: Переносите строки для условий (IF...THEN...END) или длинных выражений.  
* **Называйте поля понятно**: Это поможет быстро понять, что делает формула.  
* **Тестируйте в визуализациях**: Создайте таблицу или график, чтобы проверить, как формула работает.  
  * **Пример визуализации**: Постройте столбчатую диаграмму с полем "Profit_Calculation" ([Sales] - [Cost]), чтобы увидеть прибыль по категориям.

### Автодополнение и проверка синтаксиса
**Теория
Что такое автодополнение и проверка синтаксиса?**  
* **Автодополнение**: Инструмент в редакторе вычислений Tableau, который подсказывает названия полей (например, [Sales]), функций (например, SUM) и операторов, когда вы начинаете печатать.  
* **Проверка синтаксиса**: Tableau автоматически анализирует формулу и подсвечивает ошибки красным, если что-то не так, помогая исправить их до сохранения.

⠀**Зачем это нужно?**  
* Автодополнение ускоряет написание формул и предотвращает ошибки, такие как опечатки в названиях полей.  
* Проверка синтаксиса позволяет сразу увидеть, правильно ли написана формула, и избежать проблем при создании визуализаций.

⠀**Как это работает?**  
* **Автодополнение**:  
  * Откройте редактор вычислений (Create Calculated Field).  
  * Начните печатать, например, "SU" для SUM или "[Sa" для [Sales].  
  * Tableau покажет список подсказок — выберите нужный вариант с помощью стрелок или клика мышью.
* **Проверка синтаксиса**:  
  * Если формула правильная, внизу редактора появится надпись "The calculation is valid".  
  * Если есть ошибка (например, пропущена скобка), формула подсветится красным, и Tableau укажет проблему, например, "Expected closing parenthesis".

⠀**Типичные ошибки и как их избежать**:  
* Опечатка в названии поля: Например, [Sale] вместо [Sales].  
* Неправильный тип данных: Например, использование SUM для строк вместо чисел.  
* Синтаксические ошибки: Пропущенные скобки, лишние операторы или неправильный порядок.

⠀**Примеры**
* **Использование автодополнения**  
  * **Действие**: В редакторе начните печатать "SU".  
  * **Результат**: Tableau предложит SUM, SUBSTRING и другие функции. Выберите SUM, затем введите "[Sa", чтобы выбрать [Sales], и получите SUM([Sales]).  
  * **Что делает**: Суммирует продажи.  
  * **Результат**: Если в датасете продажи 200, 300 и 500, то SUM([Sales]) = 1000.
* **Проверка синтаксиса (правильная формула)**  
  * **Формула**:   SUM([Sales]) / COUNT([Order ID]) // Средняя сумма заказа
  * **Результат**: Tableau подтверждает "The calculation is valid".  
  * **Что делает**: Делит сумму продаж на количество заказов.
* **Проверка синтаксиса (ошибка)**  
  * **Формула**:   SUM([Sales] / COUNT([Order ID]) /* Ошибка: пропущена скобка */
  * **Результат**: Tableau подсветит формулу красным и покажет ошибку "Expected closing parenthesis".  
  * **Исправление**: SUM([Sales]) / COUNT([Order ID]).

⠀**Рекомендации для новичков**
* **Пользуйтесь автодополнением**: Начните печатать название поля или функции, чтобы избежать ошибок. Например, "[Sa" → выберите [Sales].  
* **Проверяйте синтаксис перед сохранением**: Если формула подсвечена красным, исправьте ошибку (например, добавьте пропущенную скобку).  
* **Тестируйте формулы**: Создайте простую таблицу, чтобы проверить результат формулы.  
  * **Пример визуализации**: Постройте линейный график с формулой SUM([Sales]) / COUNT([Order ID]), чтобы увидеть среднюю сумму заказа по месяцам.
* **Комбинируйте с комментариями**: Используйте // или /* */, чтобы пояснить формулу, даже если она написана с автодополнением.

### Почему это важно?
Форматирование формул с помощью комментариев (// и /* */) и отступов делает их понятными для вас и ваших коллег, особенно при работе с длинными или сложными вычислениями. Автодополнение и проверка синтаксиса ускоряют процесс написания формул и помогают избежать ошибок. Эти навыки — основа для создания надежных и читаемых вычислений, которые пригодятся в более сложных темах, таких как логические функции или LOD-вычисления.
